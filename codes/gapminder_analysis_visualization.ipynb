```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
life=pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0) 
population=pd.read_excel('gapminder_population.xlsx', index_col=0) 
continents = pd.read_csv('continents.csv',sep=';')
fert.columns = fert.columns.astype(int)
life.columns = life.columns.astype(int)
population.columns = population.columns.astype(int)
```

```python
fert = fert.reset_index()
life = life.reset_index()
population = population.reset_index()
```

```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility')
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
population = population.melt(id_vars='country', var_name='year', value_name='population')
```

```python
df=pd.merge(fert,population,how='left',on=('country','year'))
```

```python
df.head()
```

```python
df=pd.merge(df,life,how='left',on=('country','year'))
```

```python
df.head()
```

```python
df=pd.merge(df,continents,how='outer',on='country')
```

```python
df.dropna(inplace=True)
```

```python
df_subset = df.loc[df['year'] == 2000]
sns.scatterplot(x='life_expectancy', y='fertility',data=df_subset, alpha=0.6);
```

```python
df_correlation = df.loc[df['year'].between(1960, 2015)]
```

```python
sizeZZ=widgets.FloatSlider(
    value=1000000,
    min=0,
    max=1000000,
    step=100,
    description='sizes:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)

alphaZZ=widgets.FloatSlider(
    value=0.5,
    min=0,
    max=1,
    step=0.05,
    description='alpha:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)

styleZZ=widgets.Dropdown(
    options=[ 'normal' , 'italic' ,'oblique' ],
    value='italic',
    description='Fontstyle:',
    disabled=False,
)
display(sizeZZ)
display(alphaZZ)
display(styleZZ)

```

```python
x_min, x_max, = df_correlation['life_expectancy'].min(),df_correlation['life_expectancy'].max()
y_min, y_max =df_correlation['fertility'].min(),df_correlation['fertility'].max()

min_size, max_size = df["population"].min()/sizeZZ.value,df["population"].max()/sizeZZ.value
```

```python
for year in df_correlation['year'].unique():
    ax=sns.scatterplot(x='life_expectancy', y='fertility',data=df_correlation.loc[df_correlation['year'] == year],
                    alpha=alphaZZ.value,size='population',sizes=(min_size, max_size), hue='continent')
    
    plt.style.use('dark_background')
    plt.axis([x_min, x_max, y_min, y_max])
    
    plt.tick_params(axis="x", labelsize=8, labelrotation=-60,grid_alpha=0.9) #shaep
    plt.tick_params(axis="y", labelsize=8, labelrotation=-60,grid_alpha=0.2)#shaep
    
    plt.xlabel('Life Expectancy',style=styleZZ.value,alpha=0.8,fontsize=12) 
    plt.ylabel('Fertility',style=styleZZ.value,alpha=0.8,fontsize=12)
    #plt.figsize=(12,8) 
    
    plt.text(70, 8, year, fontsize=14,style='oblique',alpha=alphaZZ.value)#spice
    plt.frameon=False
    
    ax.spines['top'].set_visible(False)#remove
    ax.spines['right'].set_visible(False)#remove
    
    plt.legend('',frameon=False) #remove
    
    plt.savefig(f'lifeexp_{year}.png',dpi=800)# pixal
    plt.close()
```
